```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = np.random.randn(500,2) + np.array([0,2])
X2 = np.random.randn(500,2) + np.array([2,-2])
```

```python
X = np.vstack((X1,X2))
```

```python
X.shape
```

```python
Y = np.zeros((1000,1))
```

```python
Y[:500,:] = 0
Y[500:,:] = 1
```

```python
Y.shape
```

```python
plt.scatter(X1, X2, color=['red','blue'])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X,Y)
```

```python
coef = clf.coef_
```

```python
coef
```

```python
def sigmoid(row, coef):
    x = 0
    for i in range(len(row)):
        x += coef[0][i] * row[i]
    return 1 / (1 + np.exp(-x))
```

```python
predictions = []
for row in X:
    y_pred = sigmoid(row, coef)
    predictions.append(round(y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y, predictions)
```
