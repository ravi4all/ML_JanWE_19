```python
import pandas as pd
import bs4
import urllib.request as url
```

```python
page = url.urlopen('http://ifmd.iari.res.in/ifmd/met_data.php')
```

```python
data = bs4.BeautifulSoup(page, 'lxml')
```

```python
table = data.find('table')
```

```python
datalist = table.text.split("\n")
```

```python
# datalist[:40]
data = []
for i in range(len(datalist)):
    if datalist[i] != '':
        data.append(datalist[i])
```

```python
df = []
for i in range(len(data)):
    dataset = []
    for j in range(13*i, (13 * i) + 13):
       dataset.append(data[j])
    df.append(dataset)
```
